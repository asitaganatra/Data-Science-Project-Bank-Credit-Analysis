```python
import pandas as pd
data = pd.read_csv('cleaned_credit_data.csv') 
new_column_names = {
    'Year (year)': 'year',
    'Region (region)': 'region',
    'State Name (state_name)': 'state_name',
    'State Code (state_code)': 'state_code',
    'District Name (district_name)': 'district_name',
    'District Code (district_code)': 'district_code',
    'Population Group (population_group)': 'population_group',
    'Bank Group (bank_group)': 'bank_group',
    'Occupation Group (occupation_group)': 'occupation_group',
    'Occupation Sub Group (occupation_sub_group)': 'occupation_sub_group',
    'Number of Accounts (no_of_accounts)': 'no_of_accounts',
    'Credit Limit (credit_limit)': 'credit_limit',
    'Amount Outstanding (amount_outstanding)': 'amount_outstanding'
}
data.rename(columns=new_column_names, inplace=True)
data.head()
```

```python
#removing outliers
```

```python
# Print the original shape of the DataFrame
print(f"Original shape of the dataset: {data.shape}")

# List of numerical columns to check for outliers
numerical_cols = ['no_of_accounts', 'credit_limit', 'amount_outstanding']

# Create a copy to avoid modifying the original data directly
data_no_outliers = data.copy()

for col in numerical_cols:
    Q1 = data_no_outliers[col].quantile(0.25)
    Q3 = data_no_outliers[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter the DataFrame to keep only the rows within the bounds
    data_no_outliers = data_no_outliers[(data_no_outliers[col] >= lower_bound) & (data_no_outliers[col] <= upper_bound)]

# Print the new shape and the number of outliers removed
print(f"Shape after removing outliers: {data_no_outliers.shape}")
print(f"Number of outlier rows removed: {data.shape[0] - data_no_outliers.shape[0]}")

# Display the first few rows of the new, cleaned DataFrame
data_no_outliers.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# 1. Define Feature (X) and Target (y)
# We use double brackets for X to keep it as a DataFrame
X_simple = data[['credit_limit']]
y_simple = data['amount_outstanding']

# 2. Split data into training and testing sets
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_simple, y_simple, test_size=0.2, random_state=42)

# 3. Create and train the model
simple_model = LinearRegression()
simple_model.fit(X_train_s, y_train_s)

# 4. Make predictions and evaluate performance
y_pred_s = simple_model.predict(X_test_s)
r2_s = r2_score(y_test_s, y_pred_s)
print(f"Simple Linear Regression R-squared (R): {r2_s:.4f}")

# 5. Visualize the result
plt.figure(figsize=(10, 6))
plt.scatter(X_test_s, y_test_s, color='blue', alpha=0.5, label='Actual Values')
plt.plot(X_test_s, y_pred_s, color='red', linewidth=2, label='Regression Line')
plt.title('Simple Linear Regression: Credit Limit vs. Amount Outstanding')
plt.xlabel('Credit Limit')
plt.ylabel('Amount Outstanding')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler # Import the scaler

# --- Step 1: Feature Engineering (Same as before) ---
data['credit_limit_safe'] = data['credit_limit'].replace(0, 1e-6) 
data['utilization_ratio'] = data['amount_outstanding'] / data['credit_limit_safe']
data['high_utilization'] = (data['utilization_ratio'] > 0.75).astype(int)

# --- Step 2: Prepare and Split Data (Same as before) ---
X_class = X_encoded 
y_class = data['high_utilization']

X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_class, y_class, test_size=0.2, random_state=42, stratify=y_class)


# --- Step 3: Scale the Features ---
# Create a scaler object
scaler = StandardScaler()

# Fit the scaler ONLY on the training data and transform it
X_train_scaled = scaler.fit_transform(X_train_c)

# Transform the test data using the same scaler
X_test_scaled = scaler.transform(X_test_c)


# --- Step 4: Train the Model on SCALED Data ---
log_model = LogisticRegression(max_iter=1000, random_state=42)
# Train on the scaled data
log_model.fit(X_train_scaled, y_train_c)


# --- Step 5: Evaluate the Model ---
# Predict on the scaled data
y_pred_c = log_model.predict(X_test_scaled)

print(f"Accuracy: {accuracy_score(y_test_c, y_pred_c):.4f}\n")
print("Confusion Matrix:")
print(confusion_matrix(y_test_c, y_pred_c))
print("\nClassification Report:")
print(classification_report(y_test_c, y_pred_c))
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# --- Step 1: Select and Scale the Data ---
# We'll cluster based on the core numerical features
features_for_clustering = ['no_of_accounts', 'credit_limit', 'amount_outstanding']
X_cluster = data[features_for_clustering]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)


# --- Step 2: Use the Elbow Method to find the best 'k' ---
inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto') # Set n_init to avoid future warnings
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()


# --- Step 3: Run K-Means with the chosen k ---
# Look at the plot to find the "elbow". Let's assume it's at k=3 or k=4.
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
data['cluster'] = kmeans.fit_predict(X_scaled)

print(f"\nData successfully grouped into {optimal_k} clusters.")
print(data['cluster'].value_counts())


# --- Step 4: Analyze the Clusters ---
cluster_analysis = data.groupby('cluster')[features_for_clustering].mean()
print("\nCluster Analysis (Averages):")
print(cluster_analysis)
```

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# --- Step 1: Prepare the Data ---
# We use the same scaled data from our previous K-Means step
# X_scaled should already be defined.

# --- Step 2: Build and Train the Autoencoder ---
input_dim = X_scaled.shape[1]
latent_dim = 2 # The size of our compressed "summary"

# Define the encoder
encoder = keras.Sequential([
    keras.layers.Dense(16, activation='relu', input_shape=[input_dim]),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(latent_dim, activation='relu') # The bottleneck layer
])

# Define the decoder
decoder = keras.Sequential([
    keras.layers.Dense(8, activation='relu', input_shape=[latent_dim]),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(input_dim) # No activation on the output layer
])

# Combine into the full autoencoder model
autoencoder = keras.Sequential([encoder, decoder])

# Compile the autoencoder
autoencoder.compile(optimizer='adam', loss='mse')

# Train the model to reconstruct its own input
print("--- Training the Autoencoder... ---")
autoencoder.fit(X_scaled, X_scaled, epochs=50, batch_size=16, verbose=0)
print("Training complete.")


# --- Step 3: Use the Encoder to Create Latent Features ---
latent_features = encoder.predict(X_scaled)


# --- Step 4: Run K-Means on the New Latent Features ---
optimal_k = 3 # From our previous elbow method
kmeans_nn = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
data['autoencoder_cluster'] = kmeans_nn.fit_predict(latent_features)

print(f"\nData successfully grouped into {optimal_k} clusters using the autoencoder.")
print(data['autoencoder_cluster'].value_counts())


# --- Step 5: Analyze and Visualize the New Clusters ---
features_for_clustering = ['no_of_accounts', 'credit_limit', 'amount_outstanding']
print("\nAutoencoder Cluster Analysis (Averages):")
print(data.groupby('autoencoder_cluster')[features_for_clustering].mean())


# Visualize the new clusters
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data, 
                x='credit_limit', 
                y='amount_outstanding', 
                hue='autoencoder_cluster', 
                palette='viridis', 
                s=100,
                alpha=0.7)
plt.title('Customer Segments Found by Autoencoder + K-Means')
plt.xlabel('Credit Limit')
plt.ylabel('Amount Outstanding')
plt.legend(title='AE Cluster')
plt.grid(True)
plt.show()
```
